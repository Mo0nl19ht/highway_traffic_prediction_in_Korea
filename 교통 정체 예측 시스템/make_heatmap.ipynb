```python
import pandas as pd
import numpy as np
from tqdm import tqdm
def make_h5style(use):
    dic={}
    for k,v in tqdm( use.iterrows()):
            if int(v['링크아이디']) not in dic.keys():
                dic[int(v['링크아이디'])]=[]
            for j in v[-24:]:
              dic[int(v['링크아이디'])].append(j)
    return dic
```

```python
df=pd.read_csv(f"{path}/2017_2020_all.csv")
```

```python
dff=df.copy()
dff=dff[['일자', '요일', '도로명', '링크아이디', '시점명', '종점명', '방향', '거리', '01시', '02시','03시', '04시', '05시', '06시', '07시', '08시', '09시', '10시', '11시', '12시','13시', '14시', '15시', '16시', '17시', '18시', '19시', '20시', '21시', '22시','23시', '24시']]
a=dff[dff['도로명']=="내부순환로"]
df_up=a[a['방향']=="상행"]
```

```python
cnt=0
for date in tqdm(df_up['일자'].unique()):
        tmp=df_up[df_up['일자']==date]
        tmp2=pd.DataFrame(make_h5style(tmp))
        if cnt==0:
            cnt+=1
            arr=tmp2.values
        else:
            arr=np.concatenate([arr,tmp2.values],axis=None)

import matplotlib.pyplot as plt

hist= plt.hist(arr, bins='auto')  # arguments are passed to np.histogram

plt.show()
```

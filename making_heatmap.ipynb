```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
def make_h5style(use):
    dic={}
    for k,v in tqdm( use.iterrows()):
            if int(v['링크아이디']) not in dic.keys():
                dic[int(v['링크아이디'])]=[]
            for j in v[-24:]:
              dic[int(v['링크아이디'])].append(j)
    return dic
```

```python
df_all=[]   
for i in tqdm(range(2017,2021)):
    df=pd.read_csv(f"D:/topis_all/{i}_all.csv")
    df_all.append(df)
```

```python
for i in range(4):
    df_all[i]=df_all[i][['일자', '요일', '도로명', '링크아이디', '시점명', '종점명', '방향', '거리', '01시', '02시',
       '03시', '04시', '05시', '06시', '07시', '08시', '09시', '10시', '11시', '12시',
       '13시', '14시', '15시', '16시', '17시', '18시', '19시', '20시', '21시', '22시',
       '23시', '24시']]
```

```python
d=df_all.copy()
dff=[]
for i in range(4):
    a=d[i]
    a=a[a['도로명']=="내부순환로"]
    df_up=a[a['방향']=="상행"]
    df_up.index=range(len(df_up))
    dff.append(df_up)
```

```python
def check(down,up):
    max=85
    min=55
    
    if up > max:
        if down >max:
            #free
            return 0
        elif down <min:
            #back of queue
            return 1
        else:
            #하류 속도 50~90
            return 2
    elif up <min:
        if down > max:
            #bottleneck:
            return 3
        
        elif down <min:
            #congest
            return 4
        else:
            #하류 50~90
            return 5
    #상류 90~50
    else:
        return 6
```

```python
for df in tqdm(dff):
    for date in tqdm(df['일자'].unique()):
        tmp=df[df['일자']==date]
        tmp2=pd.DataFrame(make_h5style(tmp))
        new_df=[]
        #분류 위한 새로운 df만들기
        for rr in (tmp2.iterrows()):
            new_row=[]
            row=rr[1].tolist()
            for i,r in enumerate(row):
                if i==len(row)-1:
                    new_row.append(check(row[i-1],row[i])) 
                else:
                    new_row.append(check(r,row[i+1]))
            new_df.append(new_row)
            
                # plt.figure(figsize=(1.76,3.02), dpi=100)
        plt.imshow(new_df, cmap = 'gist_rainbow')
    #         plt.xticks(range(len(df.columns)), df.columns,rotation=30)
    #         plt.yticks(range(len(df.index)), df.index)
        plt.axis('off')
        plt.clim(0,6)
        plt.savefig(f"D:/2017_2020_내부순환_classification_img/{date}_내부순환_classification.png",dpi=100,bbox_inches='tight', pad_inches=0)
        plt.show()
        plt.close()
        
```
